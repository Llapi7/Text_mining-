```python


from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer


import matplotlib.pyplot as plt
from nltk.probability import FreqDist
```

```python
# Tokenize 

tokenized_words = word_tokenize(text)
print(tokenized_words)
```

```python
freqdist = FreqDist(tokenized_words)
print(freqdist)
```

```python
freqdist.most_common(25)
```

Tenemos el problema de que ha separado los puntos, corchetes i las comas com si fueran palabras. 
Además las mayúsculas nos contabilizan x 2 algunas palabras..

```python
freqdist.plot(40, cumulative=False)
plt.show()
```


Para solucionar esto aplicamos una list comprehension a la tokenización . Aplicaremos el método Lower() para pasarlo todo a minúsculas i tmbién el método isalpha( ) para retener solo datos numéricos.
Calcularemos de nuevo LA FREQUENCY 

```python


tokenized_words_strip = [w.lower() for w in tokenized_words if w.isalpha()]
tokenized_words_strip_dist = FreqDist(tokenized_words_strip)
print(tokenized_words_strip)

```

```python
tokenized_words_strip_dist.most_common(25)
```

# EX 2 

```python
# Remove stop words

filtered_words = [w for w in tokenized_words_strip if w not in stopwords.words('english')]
print(filtered_words)

```

```python
# buscamos la raiz de cada palabra

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print(stemmed_words)
```

```python
lemmatized_words_dist = FreqDist(stemmed_words)
lemmatized_words_dist.plot(40, title = 'word frequency distribution (whole text)', cumulative=False)
plt.show()

```
